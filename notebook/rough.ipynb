```python
import pandas as pd
```

```python
lstm_df_final = pd.read_csv('dataset_preparation/LSTM_predictions.csv')
lstm_df_final.head()
```

```python
gru_df_final = pd.read_csv('dataset_preparation/GRU_predictions.csv')
gru_df_final.head()
```

```python

# Dropping column 'B'
gru_df_final = gru_df_final.drop('dir', axis=1)
```

```python
gru_df_final.head()
```

```python
# Print the first few rows to inspect the structure of 'predicted_values'
print(lstm_df_final['predicted_values'].head().tolist())
```

```python
# Print the first few rows to inspect the structure of 'predicted_values'
print(gru_df_final['predicted_values'].head().tolist())
```

```python
import ast
```

```python

# Function to flatten a list of lists
def flatten_list(lst):
    return [item for sublist in lst if isinstance(sublist, list) for item in sublist]

# Convert string representations to actual lists and then flatten them
lstm_df_final['predicted_values'] = lstm_df_final['predicted_values'].apply(
    lambda x: flatten_list(ast.literal_eval(x))
)
```

```python
lstm_df_final.head()
```

```python
# Save the DataFrame
lstm_df_final.to_csv('dataset_preparation/lstm_df_final.csv', index=False) 
```

```python
# Convert string representations to actual lists and then flatten them
gru_df_final['predicted_values'] = gru_df_final['predicted_values'].apply(
    lambda x: flatten_list(ast.literal_eval(x))
)
```
